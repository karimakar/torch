```python
# feature Scaling
df_test = df[df['Purchase'].isnull()]
```

```python
df_test.head(2)
```

```python
df_train=df[~df['Purchase'].isnull()]
```

```python
df_train.head(2)
```

```python
y = df_train['Purchase']
y.shape
```

```python
X=df_train.drop('Purchase',axis=1)
X.shape
```

```python
# feature scaler
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=45)
```

```python
sns.scatterplot(data=df, x='Product_Category_1', y='Purchase',
                color="black", alpha=0.5)
_ = plt.title("Illustration of the regression dataset used")
```

```python
 import numpy as np
###### Single Decision Tree Regression in Python #######
from sklearn import tree
#choose from different tunable hyper parameters
RegModel = tree.DecisionTreeRegressor()
 
#Printing all the parameters of Decision Tree
print(RegModel)
 
#Creating the model on Training Data
DTree=RegModel.fit(X_train,y_train)
prediction=DTree.predict(X_test)
Predictors=['Gender','Age','Occupation','Stay_In_Current_City_Years','Marital_Status','Product_Category_1',
         'Product_Category_2','Product_Category_3','B','C']
#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value:',metrics.r2_score(y_train, DTree.predict(X_train)))
 
#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((y_test - prediction) / y_test)) * 100))
 
#Plotting the feature importance for Top 10 most important columns
%matplotlib inline
feature_importances = pd.Series(DTree.feature_importances_,index=Predictors)
feature_importances.nlargest(10).plot(kind='barh')
```

```python
 from sklearn.linear_model import LinearRegression, Ridge
#choose from different tunable hyper parameters
Lr = LinearRegression()
#Printing all the parameters of LinearRegression
print(Lr)
#Creating the model on Training Data
Alr=Lr.fit(X_train,y_train)
prediction=Alr.predict(X_test)

#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value:',metrics.r2_score(y_train, Alr.predict(X_train)))
 
#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((y_test - prediction) / y_test)) * 100))
```

```python
#choose from different tunable hyper parameters
Rg = Ridge()
#Printing all the parameters of Ridge
print(Rg)
#Creating the model on Training Data
Arg=Rg.fit(X_train,y_train)
prediction=Arg.predict(X_test)

#Measuring Goodness of fit in Training data
print('R2 Value:',metrics.r2_score(y_train, Arg.predict(X_train)))
 
#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((y_test - prediction) / y_test)) * 100))
```

## BOOSTING TECHNIQUE

```python
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor, RandomForestRegressor
#choose from different tunable hyper parameters
ada = AdaBoostRegressor()
#Printing all the parameters AdaBoostRegressor
print("the parameters AdaBoostRegressor:",ada)

#Creating the model on Training Data
print('prediction with AdaBoostRegressor ')
Ada=ada.fit(X_train,y_train)
prediction=Ada.predict(X_test)

#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value:',metrics.r2_score(y_train, Ada.predict(X_train)))
 
#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((y_test - prediction) / y_test)) * 100))
```

```python
gra = GradientBoostingRegressor()

print("the parameters GradientBoostingRegressor:",gra)

#Creating the model on Training Data
print('prediction with GradientBoostingRegressor ')
Agr=gra.fit(X_train,y_train)
prediction=Agr.predict(X_test)

#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value:',metrics.r2_score(y_train, Agr.predict(X_train)))
 
#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((y_test - prediction) / y_test)) * 100))
```

```python
ran = RandomForestRegressor()
print("the parameters RandomForestRegressor:",ran)
#Creating the model on Training Data
print('prediction with RandomForestRegressor ')
Arn=ran.fit(X_train,y_train)
prediction=Arn.predict(X_test)

#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value:',metrics.r2_score(y_train, Arn.predict(X_train)))
 
#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((y_test - prediction) / y_test)) * 100))
```
